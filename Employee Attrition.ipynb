```python
# Target Variable(Attrition)
df['Attrition'] = df['Attrition'].replace({'No': 0, 'Yes': 1})
```

```python
# Encode binary variables
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df['OverTime'] = df['OverTime'].map({'Yes': 0, 'No': 1})
```

```python
# Encode categorical columns
from sklearn.preprocessing import LabelEncoder
encoding_cols = ['BusinessTravel', 'Department',
                 'EducationField', 'JobRole', 'MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df.head()
```

```python
df.info()
```

```python
# Defining X and Y variables
x = df.drop(['Attrition'], axis=1)
y = df['Attrition'].values
```

```python
# Resampling
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
```

```python
Counter(y)
```

```python
a = RandomOverSampler(random_state=42)
x_over, y_over = a.fit_resample(x, y)
print(Counter(y_over))
```

```python
# Splitting the data into Training and Testing set
from sklearn.model_selection import train_test_split, GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(
    x_over, y_over, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
prediction = logreg.predict(x_test)
f'accuracy score:{accuracy_score(y_test,prediction):.3f}'
```

```python
# Hyperparameter Tuning
param_grid = [
    {'penalty': ['l1', 'l2'],
     'C': np.logspace(-4, 4, 20),
     'solver': ['lbfgs', 'liblinear', 'sag', 'saga'],
     'max_iter': [100, 1000, 2500, 5000]
     }
]
```

```python
clf = GridSearchCV(logreg, param_grid=param_grid,
                   cv=3, scoring='accuracy', n_jobs=-1)
best_clf = clf.fit(x_train, y_train)
```

```python
best_clf.best_estimator_
```

```python
f'accuracy score: {best_clf.score(x,y):.3f}'
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

f'accuracy score:{accuracy_score(y_test,y_pred):.3f}'
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc = rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)

f'accuracy score:{accuracy_score(y_test,y_pred):.3f}'
```
